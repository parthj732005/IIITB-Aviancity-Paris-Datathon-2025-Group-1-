## Using OpenAI's GPT model for sentiment analysis 

### 🔍 Short Summary of Code Functionality

This script analyzes Amazon product reviews by:

#### ✅ Using GPT-4 for Sentiment Classification
- Each review is classified as **Positive**, **Neutral**, or **Negative** using GPT-4 for better contextual understanding.

#### ⚠️ Detecting Misleading Reviews
Flags reviews as **misleading** when the sentiment and rating do not align:
- **Positive sentiment** with **low rating (1–2)**
- **Negative sentiment** with **high rating (4–5)**
- **Neutral sentiment** with **non-3 rating**

#### 🧾 Creates Two Columns
- `misleading_type`: Describes the type of mismatch.
- `misleading`: Boolean flag indicating whether a review is misleading.

#### 📊 Calculates and Prints Statistics
- Displays the **percentage distribution** of misleading review types.
- Specifically highlights **"Negative sentiment with high rating"** as **misleadingly positive reviews**.


````python
import openai

# Initialize OpenAI client
client = openai.OpenAI(api_key="")

# Using OpenAI's GPT model for sentiment analysis to achieve higher accuracy and better contextual understanding compared to traditional methods like VADER
def sentimentCategory(text):
    if isinstance(text, float):  # skip NaN
        return 'Neutral'
    
    try:
        prompt = f"""
        Analyze the sentiment of the following product review. 
        Classify it as one of: Positive, Neutral, or Negative.

        Review: "{text}"

        Just reply with one word: Positive, Neutral, or Negative.
        """
        
        response = client.chat.completions.create(
            model="gpt-4",
            messages=[
                {"role": "system", "content": "You classify product review sentiment accurately."},
                {"role": "user", "content": prompt}
            ],
            temperature=0
        )
        
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"Error: {e}")
        return 'Neutral'

# Apply sentiment analysis
df['SentimentCategory'] = df['fullReview'].apply(sentimentCategory)


# Define misleading conditions
conditions = [
    # Positive sentiment but low rating (1-2)
    (df['SentimentCategory'] == 'Positive') & (df['rating'].isin([1, 2])),
    # Negative sentiment but high rating (4-5)
    (df['SentimentCategory'] == 'Negative') & (df['rating'].isin([4, 5])),
    # Neutral sentiment but rating not 3
    (df['SentimentCategory'] == 'Neutral') & (df['rating'] != 3)
]

# Labels for misleading types
choices = [
    'Positive sentiment with low rating',
    'Negative sentiment with high rating',
    'Neutral sentiment with non-3 rating'
]

# Create misleading_type column
df['misleading_type'] = np.select(conditions, choices, default='Not misleading')

# Create misleading column (True if any of the conditions is met)
df['misleading'] = df['misleading_type'] != 'Not misleading'


# Calculate percentages
total_reviews = len(df)
misleading_stats = df['misleading_type'].value_counts(normalize=True) * 100


# Specific percentage for "Negative sentiment with high rating"
misleading_positive_pct = misleading_stats.get('Negative sentiment with high rating', 0)

print("\nMisleading Reviews Distribution (%):" )
print(misleading_stats)
print(f"\nPercentage of misleadingly positive reviews (Negative sentiment with high rating): {misleading_positive_pct:.2f}%")


````

# Extracting key aspects from the review along with their associated sentiment (positive, negative, or neutral)


### 🧠 Aspect-Level Sentiment Analysis Workflow Summary

#### 🔐 1. Initialization
- Sets up the OpenAI GPT-4 API client.
- Prepares a function to extract **product aspects** and their **sentiment** (Positive, Neutral, Negative) from each review.

#### 📄 2. Review Analysis
- Applies the GPT function to the first 10 reviews in the DataFrame `df['fullReview']`.
- Each review is converted into a **list of dictionaries**, where each dictionary contains:
  - `aspect`
  - `sentiment`

#### 📊 3. Aspect and Sentiment Aggregation
- Counts how often each product aspect is mentioned.
- Tracks the number of **positive**, **negative**, and **neutral** mentions for each aspect.
- Stores this data in Python dictionaries and DataFrames.

#### 📈 4. Create Final Data Table
- Builds a master table `final_df` containing:
  - Total mention count per aspect.
  - Sentiment counts per aspect.
  - A derived **satisfaction score**: `positive - negative`.

- Also creates:
  - `satisfaction_df`: Sorted aspects with **highest satisfaction**.
  - `frustration_df`: Sorted aspects with **lowest satisfaction**.

#### 💾 5. Save Results
- Saves processed data to the following CSV files:
  - `aspect_sentiment_openai1.csv`: Main DataFrame with extracted data.
  - `aspect_mention_volume1.csv`: Frequency count of each aspect.
  - `aspect_satisfaction_drivers.csv`: Ranked by satisfaction score.
  - `aspect_frustration_drivers.csv`: Ranked by frustration score.

#### 🔁 6. Fix Incomplete Responses
- Identifies rows where GPT-4 returned **malformed or empty responses**.
- Re-applies GPT extraction only for those specific rows to ensure data quality.

#### 📊 7. Visualizations
- **Bar Chart**: Shows how frequently each aspect is mentioned.
- **Pie Chart**: Displays the proportion of Positive, Negative, and Neutral sentiments overall.
- **Text Cloud Plot**: Lists aspects under each sentiment category using **color-coded labels**, grouped by sentiment type instead of traditional word cloud size.


````python
import openai
import pandas as pd
import ast
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# ------------------------------
# 1. Set API Key
# ------------------------------
client = openai.OpenAI(api_key="")
# ------------------------------
# 2. GPT-based Aspect Sentiment
# ------------------------------
def gpt_aspect_sentiment_analysis(text):
    prompt = f"""
    Extract the key aspects and their sentiment (positive, negative, neutral) from this review:

    Review: "{text}"

    Respond in this format:
    [{{"aspect": "battery", "sentiment": "positive"}}, {{"aspect": "screen", "sentiment": "negative"}}]
    """
    try:
        response = client.chat.completions.create(
            model="gpt-4",
            messages=[
                {"role": "system", "content": "You analyze product reviews for aspects and sentiment."},
                {"role": "user", "content": prompt}
            ]
        )
        return ast.literal_eval(response.choices[0].message.content.strip())
    except Exception as e:
        print(f"Error: {e}")
        return []

# ------------------------------
# 3. Load data
# ------------------------------
df['aspect_sentiments'] = df['fullReview'].head(10).apply(gpt_aspect_sentiment_analysis)

# ------------------------------
# 4. Count Aspects & Sentiments
# ------------------------------
aspect_counter = Counter()
aspect_sentiment_counter = Counter()
aspect_details = {}

for aspects in df['aspect_sentiments']:
    if not isinstance(aspects, list):
        continue
    for item in aspects:
        aspect = item['aspect'].lower()
        sentiment = item['sentiment'].lower()
        aspect_counter[aspect] += 1
        aspect_sentiment_counter[(aspect, sentiment)] += 1
        if aspect not in aspect_details:
            aspect_details[aspect] = {'positive': 0, 'negative': 0, 'neutral': 0}
        aspect_details[aspect][sentiment] += 1

# ------------------------------
# 5. DataFrames & Merge
# ------------------------------
aspect_freq_df = pd.DataFrame(aspect_counter.items(), columns=['aspect', 'mention_count']).sort_values(by='mention_count', ascending=False).reset_index(drop=True)

sentiment_df = pd.DataFrame(aspect_sentiment_counter.items(), columns=['aspect_sentiment', 'mention_count'])
sentiment_df[['aspect', 'sentiment']] = pd.DataFrame(sentiment_df['aspect_sentiment'].to_list(), index=sentiment_df.index)
sentiment_df = sentiment_df.drop(columns=['aspect_sentiment'])

pivot_df = sentiment_df.pivot(index='aspect', columns='sentiment', values='mention_count').fillna(0).reset_index()

final_df = pd.merge(aspect_freq_df, pivot_df, on='aspect', how='left')
for col in ['positive', 'negative', 'neutral']:
    if col not in final_df:
        final_df[col] = 0

final_df['satisfaction_score'] = final_df['positive'] - final_df['negative']

satisfaction_df = final_df.sort_values(by='satisfaction_score', ascending=False).reset_index(drop=True)
frustration_df = final_df.sort_values(by='satisfaction_score').reset_index(drop=True)

# ------------------------------
# 6. Save CSVs
# ------------------------------
df.to_csv("aspect_sentiment_openai1.csv", index=False)
aspect_freq_df.to_csv("aspect_mention_volume1.csv", index=False)
satisfaction_df.to_csv("aspect_satisfaction_drivers.csv", index=False)
frustration_df.to_csv("aspect_frustration_drivers.csv", index=False)
# ---------------------------------
# 7. Checking for incomplete rows and running  GPT-based Aspect Sentiment again for the same
# ---------------------------------

def is_invalid(row):
    return not isinstance(row, list) or not all(isinstance(item, dict) and 'aspect' in item and 'sentiment' in item for item in row)

invalid_indices = df[df['aspect_sentiments'].apply(is_invalid)].index

# Re-run GPT function only for missing/incomplete rows
df.loc[invalid_indices, 'aspect_sentiments'] = df.loc[invalid_indices, 'fullReview'].apply(gpt_aspect_sentiment_analysis)

# ------------------------------
# 8. VISUALIZATIONS
# ------------------------------

# Bar Chart – Aspect Mention Counts
plt.figure(figsize=(10, 6))
sns.barplot(data=final_df, x='aspect', y='mention_count', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Aspect Mention Counts')
plt.xlabel('Aspect')
plt.ylabel('Mention Count')
plt.tight_layout()
plt.show()

# Plotting pie chart for overall sentiment distribution (positive, negative, neutral)
sentiment_summary = final_df[['positive', 'negative', 'neutral']].sum()
plt.figure(figsize=(6, 6))
plt.pie(sentiment_summary, labels=sentiment_summary.index, autopct='%1.1f%%', startangle=90, colors=['#00FF00', '#FF0000', '#808080'])
plt.title('Overall Sentiment Distribution')
plt.show()
# ---------------------------------
# 9. Aspect Cloud by Sentiment Type
# ---------------------------------
aspect_data = []
for row in df['aspect_sentiments']:
    if isinstance(row, list):
        for entry in row:
            aspect_data.append({
                'aspect': entry['aspect'].lower(),
                'sentiment': entry['sentiment'].capitalize()
            })

aspect_df = pd.DataFrame(aspect_data)
grouped = aspect_df.groupby('sentiment')['aspect'].apply(list).to_dict()
grouped = {k: list(set(v)) for k, v in grouped.items()}

sentiment_order = ['Positive', 'Neutral', 'Negative']
colors = {'Positive': 'skyblue', 'Neutral': 'orange', 'Negative': 'lightcoral'}

fig, ax = plt.subplots(figsize=(16, 12))
max_len = max(len(v) for v in grouped.values())
x_positions = range(len(sentiment_order))

for i, sentiment in enumerate(sentiment_order):
    aspects = grouped.get(sentiment, [])
    for j, aspect in enumerate(aspects):
        ax.text(i, max_len - j, aspect, fontsize=10, ha='center',
                bbox=dict(boxstyle="round,pad=0.3", facecolor=colors[sentiment], edgecolor='gray'))

ax.set_xticks(x_positions)
ax.set_xticklabels(sentiment_order)
ax.set_xlim(-0.5, len(sentiment_order) - 0.5)
ax.set_ylim(0, max_len + 2)
ax.set_ylabel("Aspects")
ax.set_title("Aspects by Sentiment")
ax.set_yticks([])
plt.tight_layout()
plt.show()
````

 # Review Analysis: Extracting Sentiment, Emotion, and Customer Segments

### 🧠 Customer Review Insights via GPT-4

#### 🔍 1. GPT-Based Review Analysis
Each product review in the dataset (`df['fullReview']`) is processed using OpenAI's GPT-4 model. For every review, the model extracts:
- **Key product aspects** mentioned
- **Overall sentiment** (Positive / Neutral / Negative)
- **Dominant customer emotion** (e.g., joy, anger, frustration)
- **Customer segment** (e.g., Loyal Promoter, Vocal Critic)

The result is returned as a structured JSON-like dictionary and stored in a new column called `gpt_result`.

---

#### 🧾 2. Extracted Data Columns
From the `gpt_result` column, four new columns are created:
- `aspects`: List of product aspects mentioned
- `SentimentCategory`: Overall sentiment of the review
- `emotion`: Dominant emotion expressed by the customer
- `customer_segment`: Categorization of the customer (e.g., Loyal Promoter)

---

#### 📊 3. Emotion Distribution Visualization
The script visualizes how frequently each emotion is expressed using:
- A **bar chart** to show the count of reviews per emotion
- A **pie chart** to show the percentage share of each emotion

---

#### 🧑‍💼 4. Customer Segment Distribution Visualization
A **bar chart** is created to show how many customers belong to each segment (e.g., Loyal Promoter, Vocal Critic, Silent Quitter, Others).


````python
import pandas as pd
import ast
import openai
import matplotlib.pyplot as plt
from collections import Counter

client = openai.OpenAI(api_key="")
def gpt_extract_all(text):
    prompt = f"""
    From the review below, extract the following:
    1. List of key product aspects mentioned.
    2. Overall sentiment (Positive, Negative, Neutral).
    3. Dominant customer emotion (choose from: joy, sadness, anger, fear, surprise, love, frustration, neutral).
    4. Classify the customer into one of these segments:
        - Loyal Promoter (positive + joy/love)
        - Vocal Critic (negative + anger/frustration/disgust)
        - Silent Quitter (neutral/negative + sadness/fear/disappointment)
        - Others (all else)

    Respond in JSON format like:
    {{
        "aspects": ["battery", "design"],
        "sentiment": "Positive",
        "emotion": "joy",
        "segment": "Loyal Promoter"
    }}

    Review: "{text}"
    """

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You analyze product reviews for aspects, sentiment, emotion, and customer segment."},
                {"role": "user", "content": prompt}
            ]
        )
        return ast.literal_eval(response.choices[0].message.content.strip())
    except Exception as e:
        print(f"GPT error: {e}")
        return {"aspects": [], "sentiment": "Neutral", "emotion": "neutral", "segment": "Others"}



# Apply GPT extraction
df['gpt_result'] = df['fullReview'].apply(gpt_extract_all)

# Unpack results into separate columns
df['aspects'] = df['gpt_result'].apply(lambda x: x['aspects'])
df['SentimentCategory'] = df['gpt_result'].apply(lambda x: x['sentiment'])
df['emotion'] = df['gpt_result'].apply(lambda x: x['emotion'])
df['customer_segment'] = df['gpt_result'].apply(lambda x: x['segment'])

# Emotion distribution
def plot_emotion_distribution(df):
    emotions = df['emotion'].value_counts()
    emotions.plot(kind='bar', title='Emotion Distribution', color='skyblue')
    plt.xlabel("Emotion")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(8, 8))
    plt.pie(emotions, labels=emotions.index, autopct='%1.1f%%')
    plt.title("Emotion Share")
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

# Segment distribution
def plot_customer_segments(df):
    segments = df['customer_segment'].value_counts()
    segments.plot(kind='bar', title='Customer Segments', color='cornflowerblue')
    plt.xlabel("Segment")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Save and Plot
plot_emotion_distribution(df)
plot_customer_segments(df)
df.to_csv("gpt_output_reviews.csv", index=False)
````

### 🔍 Step-by-Step Breakdown: Competitor Mention Extraction Using GPT-4

#### 1. 🧠 GPT Prompt Setup
A custom prompt is crafted for GPT-4 to:
- Read each customer review.
- Extract a list of **competitor brands or product names** mentioned.
- Return an **empty list** if no competitors are mentioned.

---

#### 2. ⚙️ Function Execution
The function `extract_competitor_mentions()`:
- Sends the review to GPT-4.
- Parses the model's output using `ast.literal_eval()` to safely convert the string response into a Python list.

---

#### 3. 📄 DataFrame Update
- Adds a new column `mentioned_competitors` to store the list of competitor names for each review.
- Adds a boolean column `mentions_competitor` that is **`True`** if any competitors are mentioned, else **`False`**.

---

#### 4. 💾 Result Export
- The updated DataFrame is now ready to be saved or further analyzed.
- A print message confirms that the process is complete, although the actual CSV export line is not included in the code shown.


````python
import openai
import pandas as pd
import ast

# Initialize OpenAI client
client = openai.OpenAI(api_key="")
def extract_competitor_mentions(text):
    prompt = f"""
    Identify if any competitor product or brand is mentioned in this review.
    Only return a list of competitor names. If none, return an empty list.

    Review: "{text}"

    Example format: ["LG", "Sony"]
    """
    try:
        response = client.chat.completions.create(
            model="gpt-4",
            messages=[
                {"role": "system", "content": "You extract competitor brand or product names from reviews."},
                {"role": "user", "content": prompt}
            ]
        )
        content = response.choices[0].message.content.strip()
        return ast.literal_eval(content)
    except Exception as e:
        print(f"Error for review: {text[:30]}...\nReason: {e}")
        return []

# Apply function
df['mentioned_competitors']  = df['fullReview'].astype(str).apply(extract_competitor_mentions)

# Boolean column for easy filtering
df['mentions_competitor'] = df['mentioned_competitors'].apply(lambda x: len(x) > 0)

# Save results
print("Saved to reviews_with_competitor_mentions.csv")
````

# Top Complaint Detection and Sentiment Analysis Using GPT-4

### 🧾 Step-by-Step Summary: Complaint Extraction and Ranking Using GPT-4

#### ✅ Step 1: OpenAI GPT-4 Initialization
- Sets up the GPT-4 client to analyze customer review text.

---

#### 🧠 Step 2: Define Complaint Extraction Logic
For each review, GPT-4:
- Extracts a list of **specific complaints** or problem areas.
- Tags each complaint with:
  - **Sentiment**: Positive / Neutral / Negative
  - **Emotion**: Anger / Frustration / Sadness / Disappointment / Neutral
- Responds strictly in a **JSON array** format for consistency and machine readability.

---

#### 🛠️ Step 3: Apply Extraction to Reviews
- Converts all reviews to string format.
- Applies the `extract_complaints_openai()` function.
- Stores structured results in a new column: `complaint_details`.

---

#### 📂 Step 4: Flatten and Aggregate
- Extracts individual complaint entries from each review into a flat structure.
- Builds a new DataFrame `complaint_df` where each row represents one complaint instance.

---

#### 📊 Step 5: Rank Complaint Issues
- Groups complaints by text (e.g., `"handle broke"`, `"slow shipping"`).
- Computes:
  - **Frequency** of each unique complaint.
  - Most frequent **sentiment** and **emotion** associated with each complaint.

---

#### 🥇 Step 6: Get Top 3 Complaint Issues
- Identifies and ranks the **top 3 most frequently mentioned** complaints.
- Displays the result and saves it to a CSV file named:  
  `top_3_complaints_openai.csv`.


````python
import openai, json
import pandas as pd
import ast
from collections import Counter

# Step 1: Initialize OpenAI
client = openai.OpenAI(api_key="")

def extract_complaints_openai(text):
    prompt = f"""
From the following product review, extract all complaint aspects, each with:
  - sentiment: "positive", "negative", or "neutral"
  - emotion: one of "anger", "frustration", "sadness", "disappointment", "neutral"

Review:
\"\"\"{text}\"\"\"

**IMPORTANT**:
• Do NOT ask any follow-up questions.  
• Respond *only* with a JSON array of objects.  
• Example output (exactly like this, with double-quotes):
[
  {{
    "complaint": "handle broke",
    "sentiment": "negative",
    "emotion": "frustration"
  }},
  {{
    "complaint": "shipping delay",
    "sentiment": "negative",
    "emotion": "anger"
  }}
]
"""
    resp = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system",  "content": "You are a JSON-only complaint extractor."},
            {"role": "user",    "content": prompt}
        ]
    )
    content = resp.choices[0].message.content.strip()
    print("RAW MODEL OUTPUT>", repr(content))  # for debugging

    # Quick sanity check: if it doesn’t start like JSON, just return empty
    if not content.startswith('['):
        return []

    # strip code fences if any
    if content.startswith("```"):
        content = content.strip("```").strip()

    try:
        return json.loads(content)
    except json.JSONDecodeError as e:
        print("JSON parsing failed:", e)
        return []




df['fullReview'] = df['fullReview'].astype(str)

# Step 4: Apply the complaint extraction
df['complaint_details'] = df['fullReview'].apply(extract_complaints_openai)

# Step 5: Flatten complaint entries
records = []
for row in df['complaint_details']:
    if isinstance(row, list):
        for entry in row:
            records.append(entry)

complaint_df = pd.DataFrame(records)

# Step 6: Aggregate and rank complaints
ranked = (
    complaint_df
    .groupby('complaint')
    .agg( 
        frequency=('complaint', 'count'),
        emotion_mode=('emotion', lambda x: x.mode()[0] if not x.mode().empty else 'neutral'),
        sentiment_mode=('sentiment', lambda x: x.mode()[0] if not x.mode().empty else 'neutral')
    )
    .sort_values(by='frequency', ascending=False)
    .reset_index()
)

# Step 7: Save top complaints
top_3_issues = ranked.head(3)
print("Top 3 Complaint Issues:")
print(top_3_issues)

top_3_issues.to_csv("top_3_complaints_openai.csv", index=False)
````

### 🧠 GPT-4 Powered Aspect-Level Sentiment Analysis of competitve product

This notebook performs **aspect-based sentiment analysis** on product reviews using OpenAI's GPT-4, followed by aggregation and visualization. The process includes:

---

#### 🔐 1. Initialization
- Loads reviews from `conair_positive_reviews.xlsx`.
- Sets up OpenAI's GPT-4 API for analysis.

---

#### 🧠 2. Aspect Sentiment Extraction (GPT-4)
- For the first 10 reviews:
  - GPT-4 extracts **key product aspects** (e.g., "heating", "cord", "handle").
  - Classifies sentiment for each aspect as **Positive**, **Neutral**, or **Negative**.
  - Outputs a list of dictionaries per review (e.g., `[{"aspect": "cord", "sentiment": "positive"}]`).

---

#### 📊 3. Aspect & Sentiment Aggregation
- Counts total mentions of each aspect.
- Aggregates sentiment breakdown per aspect.
- Structures everything into a DataFrame called `final_df`:
  - `mention_count`: How often an aspect appears.
  - `positive`, `negative`, `neutral`: Count of sentiment mentions.
  - `satisfaction_score`: `positive - negative`.

---

#### 🥇 4. Rank Aspects by Sentiment
- Creates:
  - `satisfaction_df`: Aspects ranked by satisfaction score (high to low).
  - `frustration_df`: Aspects ranked by frustration (low to high).
  
---

#### 💾 5. Save Results
Saves key outputs to CSV:
- `aspect_sentiment_openai1.csv`
- `aspect_mention_volume1.csv`
- `aspect_satisfaction_drivers.csv`
- `aspect_frustration_drivers.csv`

---

#### 🛠️ 6. Handle Incomplete GPT Responses
- Checks for rows where GPT didn’t return valid output.
- Re-runs GPT only on those specific rows for correction.

---

#### 📊 7. Visualizations
- **Bar Chart**: Frequency of aspect mentions.
- **Pie Chart**: Distribution of positive, negative, and neutral sentiments.
- **Aspect Cloud**: Text cloud showing aspects grouped by sentiment (color-coded).



````python
import openai
import pandas as pd
import ast
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('conair_positive_reviews.xlsx')
# ------------------------------
# 1. Set API Key
# ------------------------------
client = openai.OpenAI(api_key="")
# ------------------------------
# 2. GPT-based Aspect Sentiment
# ------------------------------
def gpt_aspect_sentiment_analysis(text):
    prompt = f"""
    You are a product review analyst. Extract key aspects and their sentiment from the review below.
    
    Each aspect should be one or two words (e.g., 'battery life', 'screen', 'ironing'). Sentiment should be one of: Positive, Neutral, Negative.
    
    Make sure to consider the full context of the review — especially phrases with negation like 'no more', 'doesn’t', or 'not needed'.
    
    Review:
    \"{text}\"
    
    Respond only as a Python list of dictionaries like:
    [{{"aspect": "ironing", "sentiment": "positive"}}, ...]
    """

    try:
        response = client.chat.completions.create(
            model="gpt-4",
            messages=[
                {"role": "system", "content": "You analyze product reviews for aspects and sentiment."},
                {"role": "user", "content": prompt}
            ]
        )
        return ast.literal_eval(response.choices[0].message.content.strip())
    except Exception as e:
        print(f"Error: {e}")
        return []

# ------------------------------
# 3. Load data
# ------------------------------
# df = pd.read_csv("your_reviews.csv")  # Uncomment for actual use
df['aspect_sentiments'] = df['fullReview'].head(10).apply(gpt_aspect_sentiment_analysis)

# ------------------------------
# 4. Count Aspects & Sentiments
# ------------------------------
aspect_counter = Counter()
aspect_sentiment_counter = Counter()
aspect_details = {}

for aspects in df['aspect_sentiments']:
    if not isinstance(aspects, list):
        continue
    for item in aspects:
        aspect = item['aspect'].lower()
        sentiment = item['sentiment'].lower()
        aspect_counter[aspect] += 1
        aspect_sentiment_counter[(aspect, sentiment)] += 1
        if aspect not in aspect_details:
            aspect_details[aspect] = {'positive': 0, 'negative': 0, 'neutral': 0}
        aspect_details[aspect][sentiment] += 1

# ------------------------------
# 5. DataFrames & Merge
# ------------------------------
aspect_freq_df = pd.DataFrame(aspect_counter.items(), columns=['aspect', 'mention_count']).sort_values(by='mention_count', ascending=False).reset_index(drop=True)

sentiment_df = pd.DataFrame(aspect_sentiment_counter.items(), columns=['aspect_sentiment', 'mention_count'])
sentiment_df[['aspect', 'sentiment']] = pd.DataFrame(sentiment_df['aspect_sentiment'].to_list(), index=sentiment_df.index)
sentiment_df = sentiment_df.drop(columns=['aspect_sentiment'])

pivot_df = sentiment_df.pivot(index='aspect', columns='sentiment', values='mention_count').fillna(0).reset_index()

final_df = pd.merge(aspect_freq_df, pivot_df, on='aspect', how='left')
for col in ['positive', 'negative', 'neutral']:
    if col not in final_df:
        final_df[col] = 0

final_df['satisfaction_score'] = final_df['positive'] - final_df['negative']

satisfaction_df = final_df.sort_values(by='satisfaction_score', ascending=False).reset_index(drop=True)
frustration_df = final_df.sort_values(by='satisfaction_score').reset_index(drop=True)

# ------------------------------
# 6. Save CSVs
# ------------------------------
df.to_csv("aspect_sentiment_openai1.csv", index=False)
aspect_freq_df.to_csv("aspect_mention_volume1.csv", index=False)
satisfaction_df.to_csv("aspect_satisfaction_drivers.csv", index=False)
frustration_df.to_csv("aspect_frustration_drivers.csv", index=False)
# ---------------------------------
# 7. Checking for incomplete rows and running  GPT-based Aspect Sentiment again for the same
# ---------------------------------

def is_invalid(row):
    return not isinstance(row, list) or not all(isinstance(item, dict) and 'aspect' in item and 'sentiment' in item for item in row)

invalid_indices = df[df['aspect_sentiments'].apply(is_invalid)].index

# Re-run GPT function only for missing/incomplete rows
df.loc[invalid_indices, 'aspect_sentiments'] = df.loc[invalid_indices, 'fullReview'].apply(gpt_aspect_sentiment_analysis)

# ------------------------------
# 8. VISUALIZATIONS
# ------------------------------

# Bar Chart – Aspect Mention Counts
plt.figure(figsize=(10, 6))
sns.barplot(data=final_df, x='aspect', y='mention_count', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Aspect Mention Counts')
plt.xlabel('Aspect')
plt.ylabel('Mention Count')
plt.tight_layout()
plt.show()

# Plotting pie chart for overall sentiment distribution (positive, negative, neutral)
sentiment_summary = final_df[['positive', 'negative', 'neutral']].sum()
plt.figure(figsize=(6, 6))
plt.pie(sentiment_summary, labels=sentiment_summary.index, autopct='%1.1f%%', startangle=90, colors=['#00FF00', '#FF0000', '#808080'])
plt.title('Overall Sentiment Distribution')
plt.show()
# ---------------------------------
# 9. Aspect Cloud by Sentiment Type
# ---------------------------------
aspect_data = []
for row in df['aspect_sentiments']:
    if isinstance(row, list):
        for entry in row:
            aspect_data.append({
                'aspect': entry['aspect'].lower(),
                'sentiment': entry['sentiment'].capitalize()
            })

aspect_df = pd.DataFrame(aspect_data)
grouped = aspect_df.groupby('sentiment')['aspect'].apply(list).to_dict()
grouped = {k: list(set(v)) for k, v in grouped.items()}

sentiment_order = ['Positive', 'Neutral', 'Negative']
colors = {'Positive': 'skyblue', 'Neutral': 'orange', 'Negative': 'lightcoral'}

fig, ax = plt.subplots(figsize=(10, 8))
max_len = max(len(v) for v in grouped.values())
x_positions = range(len(sentiment_order))

for i, sentiment in enumerate(sentiment_order):
    aspects = grouped.get(sentiment, [])
    for j, aspect in enumerate(aspects):
        ax.text(i, max_len - j, aspect, fontsize=10, ha='center',
                bbox=dict(boxstyle="round,pad=0.3", facecolor=colors[sentiment], edgecolor='gray'))

ax.set_xticks(x_positions)
ax.set_xticklabels(sentiment_order)
ax.set_xlim(-0.5, len(sentiment_order) - 0.5)
ax.set_ylim(0, max_len + 2)
ax.set_ylabel("Aspects")
ax.set_title("Aspects by Sentiment")
ax.set_yticks([])
plt.tight_layout()
plt.show()
````
